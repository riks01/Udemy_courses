```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
df = pd.read_csv('../data/daily-total-female-births-CA.csv',parse_dates=[0], index_col=[0])
```

```python
plot_acf(df['births'])
```

```python
plot_pacf(df['births'])
```

```python
train = df['births'][:320]
test = df['births'][320:]
```

```python
train.shape
```

```python
test.shape
```

```python
arm = ARIMA(train,order =(2,1,3))
```

```python
model=arm.fit()
```

```python
pred = model.forecast(steps=45)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test,pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scData = scaler.fit_transform(df)
```

```python
scData
```

```python
train = scData[:320]
test = scData[320:]
arm = ARIMA(train,order=(2,1,3))
model = arm.fit()
```

```python
sc_pred = model.forecast(steps=45)
```

```python
np.sqrt(mean_squared_error(sc_pred, test))
```

```python
ac_pred = scaler.inverse_transform(sc_pred.reshape(-1,1))
```

```python
ac_pred[:5]
```

```python
np.sqrt(mean_squared_error(ac_pred, scaler.inverse_transform(test)))
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['births'])
```

```python
result
```

```python
def ad_fuller(column):
    result = adfuller(df[column])
    labels = ['adf','p-value','used-lag', 'nobs', 'critical-values', 'icbest', 'resstore' ]
    final = zip(labels, result)
    for label, value in final:
        print(f'{label} : {value}')
    if result[1] < 5:
        print('We don\'t have enough evidence to accept the null hypothesis, therefore our data is stationary ')
    else:
        print('We don\'t have enough evidence to reject the null hypothesis, therefore our data is non-stationary ')
```

```python
ad_fuller('births')
```
